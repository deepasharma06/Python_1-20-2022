```python
#1 Write a Python program to compute the eigenvalues and eigenvectors of a given square array.
from numpy import array
from numpy.linalg import eig

A = array([
[1, 0, 2, 3],
[1, 3, 0, 2],
[2, 0, 1, 3],
[0, 3, 2, 1]])
print(A)

# factorize
values, vectors = eig(A)
print ('the Eigen Values are:') 
print(values)
print('the Eigen Vector are:')
print (vectors)
```

```python
# Q2
# Define determinant of a 2 by 2 matrix
def det2x2(m):
  return m[0][0] * m[1][1] - m[0][1] * m[1][0]

# Define your martix
matrix = [
    [ 1, 2, 3],
    [ 5, 2, 9],
    [ 2, 3, 1]
]

a, b, c = matrix[0]

print("matrix:")
for array in matrix: print(array)
print()

# Define determinant of a 3 by 3 matrix

efhi = [x[1:] for  x in matrix[1:]]

dfgi = [x[::2] for x in matrix[1:]]

degh = [x[0:2] for x in matrix[1:]]        

det = (
    a * det2x2(efhi)
    -b * det2x2(dfgi)
    +c * det2x2(degh)
)
# Print Determinant

print(f'Determinant = {det}')
```

```python
#3 Write a Python program that solves a system of inequalities with a 4 x 4 square matrix A where Ax=b.
import numpy as np
from numpy import array

A = np.matrix([
[1, 2, 3,5],
[4, 5, 6,7],
[7, 8, 9,8],
[3,7,6,4]])

b = np.matrix([
[-1, 1, 4,6],
[2, 3, -2,4],
[7, 8, 9,8],
[3,7,6,4]])
np.linalg.solve(A,b)
```
